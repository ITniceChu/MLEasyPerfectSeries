```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['figure.dpi']= 120
```

# Linear Algebra 소개

```python
df = pd.read_csv('500_weight_height.csv')
```

```python
df
```

```python
tempdf = df[(df.Gender=='Male')&(df.Index==2)]

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(tempdf.Height.values.reshape(-1,1), tempdf.Weight)
pred_line = reg.predict((np.array(list(range(1500,2000)))/10).reshape(-1,1))

plt.scatter(tempdf.Height, tempdf.Weight)
plt.scatter((np.array(list(range(1500,2000)))/10),pred_line,s=3)
plt.title('Normal BMI Profiles',fontsize=14)
plt.xlabel('Height',fontsize=14)
plt.ylabel('Weight',fontsize=14)
```

```python
tempdf = df[(df.Gender=='Male')]

from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(tempdf.Height.values.reshape(-1,1), tempdf.Weight)
pred_line = reg.predict((np.array(list(range(1400,2000)))/10).reshape(-1,1))

plt.scatter(tempdf.Height, tempdf.Weight)
plt.scatter((np.array(list(range(1400,2000)))/10),pred_line,s=3)
plt.title('All BMI Profiles',fontsize=14)
plt.xlabel('Height',fontsize=14)
plt.ylabel('Weight',fontsize=14)
```
